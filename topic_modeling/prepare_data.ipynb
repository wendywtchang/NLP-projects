```python
import pandas as pd
import numpy as np
import json
import os
import re
```

<h2>Overview of data1: SARS</h2>

```python
data1 = "dataset/SARS-CoV.xml"
data2 = "dataset/MERS-CoV.xml"
data3 = "dataset/SARS-CoV-2.xml"
```

```python
dict_SARS = {'Pmid':[], 'Year':[], 'Title':[], 'Abstract':[],'Keywords':[]} 
title = []
abstract = []
year = []
pmid = []
keywords = []

count = 0

f = open(data1, "r")
for l in f:  
    if "/PMID" in l:
        s = [el.replace("\n", "") for el in re.split(r'<\/?[A-Za-z]+[\s]?[A-Za-z]*=?"?[0-9]*"?>', l)]
        #print(s)
        sent = list(filter(None, s))
        #print(sent)
        pmid.append(sent[0])
        
    if "/Year" in l:
        s = [el.replace("\n", "") for el in re.split(r'<\/?[A-Za-z]+[\s]?[A-Za-z]*=?"?[0-9]*"?>', l)]
        sent = list(filter(None, s))
        #print(sent)
        year.append(sent[0])
        
    if "/ArticleTitle" in l:
        s = [el.replace("\n", "") for el in re.split(r'<\/?[A-Za-z]+[\s]?[A-Za-z]*=?"?[0-9]*"?>', l)]
        sent = list(filter(None, s))
        #print(sent)
        title.append(sent[0])
        
    if "/AbstractText" in l:
        s = [el.replace("\n", "") for el in re.split(r'<\/?[A-Za-z]+[\s]?[A-Za-z]*=?"?[0-9]*"?>', l)]
        sent = list(filter(None, s))
        #print(sent)
        abstract.append(sent[0])
        
    if "/Keyword" in l:
        s = [el.replace("\n", "") for el in re.split(r'<\/?[A-Za-z]+[\s]?[A-Za-z]*=?"?[A-Za-z]*"?>', l)]
        sent = list(filter(None, s))
        #print(sent)
        keywords.append(sent)
        
    # end of an article    
    if "</PubmedArticle>" in l:
        dict_SARS['Title'].append(title)
        dict_SARS['Abstract'].append(abstract)
        dict_SARS['Year'].append(year[0]) # only the first one
        dict_SARS['Pmid'].append(pmid[0])
        dict_SARS['Keywords'].append(keywords)
        
        # clear space for next doc
        title = []
        abstract = []
        year = []
        pmid = []
        keywords = []
        
f.close()
#print(dict_SARS)
df_SARS = pd.DataFrame(data=dict_SARS)
df_SARS
```

```python
# output json file
df_SARS.to_json(r'dataset/SARS-CoV.json')
```

<h2>Delete empty abstract document</h2>

```python
# open title data
data_path = "dataset/"
df_corona = pd.read_json(data_path + 'corona_v3.json')
df_corona
```

```python
# number of empty abstract
count = 0
for ab in df_corona["Abstract"]:
    if ab == []:
        count += 1
print(count)
```

```python
# remove df.drop()
df_drop = df_corona
drop_ls = []
idx = 0

for ab in df_drop["Abstract"]:
    if ab == []:
        #df_drop = df_drop.drop(idx)
        drop_ls.append(idx)
        idx += 1       
    else:
        idx += 1
print(idx)
print(len(drop_ls))
```

```python
for i in drop_ls:
    df_drop = df_drop.drop(i)
df_drop
```

```python
# output json file
df_drop.to_json(r'dataset/drop.json')
```

```python
early = 0
mid = 0
late = 0
for y in df_drop["Year"]:
    if int(y) < 2002:
        early += 1
    if int(y) >= 2002 and int(y) < 2019:
        mid += 1
    if int(y) >= 2019:
        late += 1
print(early)
print(mid)
print(late)
print(early + mid + late)
        
```

```python
# open title data
data_path = "dataset/"
df_corona = pd.read_json(data_path + 'corona_v3.json')
df_corona
```

```python
data4 = "dataset/corona.xml"
```

```python
dict_month = {'Month':[]} 

month = []

count = 0

f = open(data4, "r")
for l in f: 
    if "/Month" in l:
        s = [el.replace("\n", "") for el in re.split(r'<\/?[A-Za-z]+[\s]?[A-Za-z]*=?"?[0-9]*"?>', l)]
        sent = list(filter(None, s))
        #print(sent)
        month.append(sent[0])
        
    # end of an article    
    if "</PubmedArticle>" in l:
        dict_month['Month'].append(month[0]) # only the first one
        
        # clear space for next doc
        month = []

        
f.close()
#print(dict_SARS)
df_month = pd.DataFrame(data=dict_month)
df_month
```

```python
df_corona["Month"] = df_month
df_corona
```

```python
# open title data
data_path = "dataset/"
df_dtm = pd.read_json(data_path + 'corona_dtm.json')
df_dtm
```

```python
# make date column
date = []
id = 0

for month in df_dtm["Month"]:
    string_date = str(month) + "/" + str(df_dtm["Year"][id])
    date.append(string_date)
    id += 1
    
df_dtm["Date"] = date
df_dtm
```

```python
df_dtm['Date'] =pd.to_datetime(df_dtm.Date)
df_dtm.sort_values(by='Date') # This now sorts in date order
```

```python
# output json file
df_dtm.to_json(r'dataset/corona_dtm.json')
```
